```
values = [2, 4, 7, 7]
causes = ['Undetermined', 'Powerlines', 'Human Related', 'Lightning']
my_colors = ['lightcoral', 'lightsteelblue',  'mediumturquoise', 'palegoldenrod']
my_color = ['lightcoral', 'cornflowerblue', 'lightseagreen', 'goldenrod']
my_explode = [0, 0, 0.1, 0]
```

```
plt.figure(1, figsize=(18,9)) 
the_grid = GridSpec(2, 2)
plt.pie(values, labels = causes, autopct='%1.f%%', startangle=115, colors = my_color, shadow = True, explode = my_explode)
plt.title('Cause of Fires', fontsize =20)
plt.axis('equal')
plt.show()
```
